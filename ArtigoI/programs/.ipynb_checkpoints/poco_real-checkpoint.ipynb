```python
#Pacotes
%matplotlib inline
import numpy as np
import pylab as py
import matplotlib.pyplot as plt
from matplotlib import style
import matplotlib.transforms as mtransforms
from matplotlib.patches import Polygon
import scipy as sp
import pandas as pd
#----------------------------------
```

```python
#classi = pd.read_table('class.txt', delim_whitespace=True, names=('Codigo','Depth(m)','SP','RLAT','TTI','TOT'))
classi = np.loadtxt(fname='../inputs/classificacao1API0001.txt', skiprows=2, usecols=(1,2,3,4,5,6)) 
```

```python
#armazenando os dados de classificação em um vetores
cod = classi[:,0]
prof = classi[:,1]
GR = classi[:,2]
RHOB = classi[:,3]
DT= classi[:,4]
SP = classi[:,5]
```

```python
#Gráfico com Dados Somente
#style.use('ggplot')
#plt.close('all')
#plt.style.use(['classic'])
py.rcParams['figure.figsize'] = (5.0, 20.0)#Redimensiona a figura
plt.xlim(1, 150)
plt.ylim(3989, 4030)
#plt.semilogx(,prof, 'r', label=b'Gamma ($\Omega m$)')
plt.plot(GR, prof, 'b', label=r'GR ($API$)')
plt.gca().invert_yaxis()
plt.xlabel('Raio-Gamma')
plt.ylabel('Profundidade (m)')
plt.title('$1API0001PR$', fontsize='xx-large')
plt.legend(loc='upper right')
plt.grid()
plt.savefig('../outputs/1API0001PR_res.png')
#plt.show()
```

```python
#Gráfico com Dados Somente
#plt.style.use(['classic'])
py.rcParams['figure.figsize'] = (5.0, 20.0)#Redimensiona a figura
plt.xlim(1, 3)
plt.ylim(3989, 4030)
plt.plot(RHOB, prof, 'r', label=r'RHOB ($g/cm^{3}$)')
plt.gca().invert_yaxis()
plt.xlabel('Densidade')
plt.ylabel('Profundidade (m)')
plt.title('$1API0001PR$', fontsize='xx-large')
plt.legend(loc='upper right')
plt.grid()
plt.savefig('../outputs/1API0001PR_dens.png')
#plt.show()
```

```python
#Gráfico com Dados Somente
#plt.style.use(['classic'])
py.rcParams['figure.figsize'] = (5.0, 20.0)#Redimensiona a figura
#plt.xlim(10e-1, 10e3)
plt.ylim(3989, 4030)
plt.plot(DT,prof, 'g', label='DT ($ms.m$)')
plt.gca().invert_yaxis()
plt.xlabel('Sonico')
plt.ylabel('Profundidade (m)')
plt.title('$1API0001PR$', fontsize='xx-large')
plt.legend(loc='upper right')
plt.grid()
plt.savefig('../outputs/1API0001PR_sonic.png')
#plt.show()
```

```python
#Gráfico com Dados Somente
#plt.style.use(['classic'])
py.rcParams['figure.figsize'] = (5.0, 20.0)#Redimensiona a figura
#plt.xlim(-0.5, 3.5)
plt.ylim(3989, 4030)
plt.plot(SP,prof, 'k', label='SP($mV$)')
plt.gca().invert_yaxis()
plt.xlabel('Potencial Espontâneo')
plt.ylabel('Profundidade (m)')
plt.title('$1API0001PR$', fontsize='xx-large')
plt.legend(loc='upper right')
plt.grid()
plt.savefig('../outputs.1API0001PR_sp.png')
plt.show()
```

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(nrows=1, ncols=4, sharey=True)


py.rcParams['figure.figsize'] = (10.0, 20.0)#Redimensiona a figura


ax1.set_xlim(1, 150)
ax1.set_ylim(3989, 4030)
ax1.plot(GR, prof, 'b', label=r'GR ($API$)')
ax1.invert_yaxis()
#ax1.set_title('$GR$', fontsize='xx-large')
ax1.legend(loc='upper right')
ax1.grid()

ax2.set_xlim(1, 3)
ax2.set_ylim(3989, 4030)
ax2.plot(RHOB, prof, 'r', label=r'RHOB ($g/cm^{3}$)')
ax2.invert_yaxis()
#ax2.set_title('Densidade', fontsize='xx-large')
ax2.legend(loc='upper right')
ax2.grid()


ax3.set_ylim(3989, 4030)
ax3.plot(DT,prof, 'g', label='DT ($ms.m$)')
ax3.invert_yaxis()
#ax3.set_title('Sônico', fontsize='xx-large')
ax3.legend(loc='upper right')
ax3.grid()


ax4.set_ylim(3989, 4030)
ax4.plot(SP,prof, 'k', label='SP($mV$)')
ax4.invert_yaxis()
#ax4.set_title('Potencial Espontâneo', fontsize='xx-large')
ax4.legend(loc='upper right')
ax4.grid()


fig.subplots_adjust(hspace=0.2,wspace=0.5)

plt.title('Seção parcial do poço 1API0001PR', fontsize='xx-large')
plt.draw()
plt.savefig('../outputs/1API0001PR.png')
plt.show()
plt.close()
```
