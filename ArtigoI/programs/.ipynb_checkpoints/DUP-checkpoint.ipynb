## Primeira derivada unidimensional progressiva

```python
#  Required Packages
import pylab as py
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=np.loadtxt('../outputs/dxdy_vel10%.txt')
```

```python
x=data[:,0]
dzdx=data[:,1]
```

```python
py.rcParams['figure.figsize'] = (5.0, 25.0)#Redimensiona a figura
%config InlineBackend.figure_format='retina'
fig, ax = plt.subplots(nrows=1, ncols=1, sharex=True)

ax.plot(dzdx,x, 'k.-')
ax.set_ylim([0,700])
ax.invert_yaxis()
ax.set_title('Interfaces de velocidade')
ax.set_ylabel('Profundidade x 10 (m)',fontsize=20)
ax.set_xlabel('dzdx',fontsize=20)
ax.grid(True)

plt.savefig('Velocity_contact.png')
plt.show()
```
