# Linear Discrimitation Analysis


```python
#  Required Packages
import pylab as py
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from IPython.display import Latex
import math 
from turtle import *
%matplotlib inline
```

```python
data1=np.loadtxt('../outputs/HIP_diabasio_r30%.txt',skiprows=3)
```

```python
#Lendo o dado do hiperplano.f90
x1=data1[0:8,0]
y1=data1[0:8,1]
x2=data1[8:16,0]
y2=data1[8:16,1]
```

```python
markerline, stemlines, baseline = plt.stem(x1, y1, 'k-.',markerfmt='go', label='Diabásio')#,markersize=30)
markerline, stemlines, baseline = plt.stem(x1, y2, 'k-.', markerfmt='ro', label='Não-Diabásio')#, markersize=30)


py.rcParams['figure.figsize'] = (20.0, 10.0)#Redimensiona a figura
plt.rcParams['legend.fontsize'] = 20 #tamanho da letra na legenda

plt.setp(baseline, color='k',  label='Projeção do hiperplano',linewidth=1)

plt.title("LDA Diabásio (Ruído=30%)", fontweight=None, fontsize=30)
plt.tick_params(axis='both', labelsize=25)#aumenta o tamanho dos valores nos eixos

plt.grid(True)

plt.xlabel("Dado de treinamento", fontsize=25, labelpad=3)
plt.ylabel("Distâncias", fontsize=25)
plt.annotate("$\Psi$ = 25%", 
            xy=(5, 3),
            xycoords='data',
            xytext=(1.1, 3),
            textcoords='data', fontsize=20)
            
plt.legend(shadow=True)
plt.savefig('../outputs/LDA-Diabasio30%.png')  
plt.show()  
plt.close()
plt.show()
```

O diagrama indica que 62% dos pontos usados no treinamento foram corretamente separados pelo LDA.

## Índice de Separabilidade Linear

$$\Upsilon = |\sum_{i=1}^{n} \dfrac{\delta_{i}}{|\delta_{i}|}| + |\sum_{j=1}^{m} \dfrac{\delta_{j}}{|\delta_{j}|}| $$
